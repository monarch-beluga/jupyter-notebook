```python
import cinrad
import numpy as np
import matplotlib.pyplot as plt
import geemap
import ee
```

```python
Map = geemap.Map()
Map
```

```python
path = r'F:/文件/工作/leida/'
file = path + 'Z_RADR_I_Z9250_20160701001000_O_DOR_SA_CAP.bin'
```

```python
f = cinrad.io.CinradReader(file)
```

```python
r = f.get_data(0,400,"REF")
r
```

```python
lon = r["longitude"].values.reshape(-1, 1)
lat = r["latitude"].values.reshape(-1, 1)
var = r['REF'].values.reshape(-1, 1)
geos = np.hstack([lon, lat, var])
geos = np.delete(geos,np.where(np.isnan(geos))[0],axis=0)
```

```python
def getPoint(geo, List):
    geo_s = ee.List(geo)
    feature = ee.Feature(ee.Geometry.Point([geo_s.get(0), geo_s.get(1)]))
    feature = feature.set('ref', geo_s.get(2))
    return ee.List(List).add(feature)
```

```python
palette = ['FF0000', '00FF00', '0000FF']
empty = ee.Image().byte()
geo_List = ee.List(geos.tolist()[:20000])
geo_point = ee.List([])
Geometrys = ee.FeatureCollection(ee.List(geo_List.iterate(getPoint, geo_point)))
outlines1 = empty.paint(
    featureCollection=Geometrys,
    color='ref'
)
```

```python
geo_List = ee.List(geos.tolist()[20000:40000])
geo_point = ee.List([])
Geometrys = ee.FeatureCollection(ee.List(geo_List.iterate(getPoint, geo_point)))
outlines2 = empty.paint(
    featureCollection=Geometrys,
    color='ref'
)
```

```python
geo_List = ee.List(geos.tolist()[40000:60000])
geo_point = ee.List([])
Geometrys = ee.FeatureCollection(ee.List(geo_List.iterate(getPoint, geo_point)))
outlines3 = empty.paint(
    featureCollection=Geometrys,
    color='ref'
)
```

```python
geo_List = ee.List(geos.tolist()[60000:80000])
geo_point = ee.List([])
Geometrys = ee.FeatureCollection(ee.List(geo_List.iterate(getPoint, geo_point)))
outlines4 = empty.paint(
    featureCollection=Geometrys,
    color='ref'
)
```

```python
geo_List = ee.List(geos.tolist()[80000:len(geos)])
geo_point = ee.List([])
Geometrys = ee.FeatureCollection(ee.List(geo_List.iterate(getPoint, geo_point)))
outlines5 = empty.paint(
    featureCollection=Geometrys,
    color='ref'
)
```

```python
leida_c = ee.ImageCollection([outlines1,outlines2,outlines3,outlines4,outlines5])
```

```python
leida = leida_c.mosaic()
```

```python
Map.addLayer(outlines1, {'palette': palette}, 'leida1')
Map.addLayer(outlines2, {'palette': palette}, 'leida2')
Map.addLayer(outlines3, {'palette': palette}, 'leida3')
Map.addLayer(outlines4, {'palette': palette}, 'leida4')
Map.addLayer(outlines5, {'palette': palette}, 'leida5')
```

```python
geemap.ee_export_image(outlines1, 'F:/leida1.tif', scale=1000, region=outlines1.geometry())
geemap.ee_export_image(outlines2, 'F:/leida2.tif', scale=1000, region=outlines1.geometry())
geemap.ee_export_image(outlines3, 'F:/leida3.tif', scale=1000, region=outlines1.geometry())
geemap.ee_export_image(outlines4, 'F:/leida4.tif', scale=1000, region=outlines1.geometry())
geemap.ee_export_image(outlines5, 'F:/leida5.tif', scale=1000, region=outlines1.geometry())
```
